```python
#Wikepedia Table
def scraping_wikipedia_table(url, i):
   
    response1 = requests.get(url)

    TeamName=[]
    TitleColumnName=[]
    TableList=[]
#TableList.append([])
    TitleColumnName.remove
    TableList.remove
    soup = BeautifulSoup(response1.text)
    table = soup.findAll( "table")
    alltr=table[i].findAll("tr")
    titlecolumn=alltr[0].findAll("th")
    if len(titlecolumn)>0:
        for columnname in titlecolumn:
            TitleColumnName.append(columnname.string)

    elif len(titlecolumn)==0:
        titlecolumn=alltr[0].findAll("td")
        for columnname in titlecolumn:
            TitleColumnName.append(columnname.string)
            
    alltr=table[i].findAll("tr")
    for x in range(1, len(alltr)-1):
        TableList.append([])
        findalltd=alltr[x].findAll("th")
        row=alltr[x].findAll("td")       
        for y in range(0,len(findalltd)):  
            TableList[x-1].append(findalltd[y].text)
        for y in range(0, len(row)): 
            TableList[x-1].append(row[y].text)    
    df = pd.DataFrame(data = TableList, columns=TitleColumnName)
    return df
```

```python
def scraping_walmart_movies(url):
    
    parray=[]
    marray=[]
    sarray=[]
    aarray=[]
    fdarray=[]
    fsarray=[]
    rarray=[]
    runarray=[]
    ratarray=[]
    ratingarray=[]
    formatarray=[]
    fmarray=[]

    FreeDeleveryString=''

    response1 = requests.get(url)
    soup = BeautifulSoup(response1.text)
    MovieDiv = soup.findAll('div','tile-content')

    for i in range(0,10):
        MovieName=MovieDiv[i].find('h4','tile-heading')
        P=MovieDiv[i].find('span','price price-display')
        if P is  None:
             P=MovieDiv[i].find('span','price price-display price-not-available')
        pricesplit=P.text.split("$")
            
        S=MovieDiv[i].findAll('span','visuallyhidden')
        
        if len(S) < 2:
            sarray.insert(i,np.nan)
        elif len(S) > 2:    
            rsplit=S[1].text.split(" ")
            sarray.insert(i,rsplit[0])
   
    
        F=MovieDiv[i].find('div','tile-aside-content')
        J=MovieDiv[i].find('dl','media-details dl-horizontal copy-mini')
    
        rarray=J.text.split("Running:")
        M=rarray[1].split("minutes")
        runarray.insert(i,M[0])

        ratarray=J.text.split("Rating:")
        ratingarray.insert(i,ratarray[1])
    
        if "Format" in J.text:
            fmarray=J.text.split("Format:")
            formatarray.insert(i,fmarray[1].split("Rating")[0].split("Release")[0])
            
        if "Format" not in J.text:
            formatarray.insert(i,np.nan)    
            
        if "Starring:"  in J.text:
            A=MovieDiv[i].find('dd','media-details-multi-line media-details-artist-dd module')
            aarray.insert(i,A.text)
        
        if "Starring:" not in J.text:
            aarray.insert(i,np.nan)
    
        if "Free shipping on orders over $50" not in F.text:
            fdarray.insert(i,"false")
        
        if "Free shipping on orders over $50" in F.text:
            fdarray.insert(i,"true")
        
        if "Free store pickup" not in F.text:
            fsarray.insert(i,"false")
        
        if "Free store pickup" in F.text:
            fsarray.insert(i,"true") 
        
        parray.insert(i,pricesplit[1])
        marray.insert(i,MovieName.text)
    
    PriceArray=np.array(parray)
    MovieArray=np.array(marray)
    StarArray=np.array(sarray)
    ActorArray=np.array(aarray)
    FreeDelivery=np.array(fdarray)
    FreeStore=np.array(fsarray)
    RunningTime=np.array(runarray)
    Rating=np.array(ratingarray)
    Format=np.array(formatarray)

    d = {'Movie Name' : Series(MovieArray),'Price' : Series(PriceArray),'Rating' : Series(StarArray),'Starring' : Series(ActorArray),'Free Delivery' : Series(FreeDelivery),'Free Store PickUp' : Series(FreeStore),'Running Time':Series(RunningTime),'Rating':Series(Rating),'Format':Series(Format)}
    df = DataFrame(d)
    df.fillna("nan")
    return df
```

```python
#Task 7a: Write two functions: one for Obama and one for Romney that does the following:
# Given a value of N (N can be 1, 2, 5, 10 etc), sort the contributions made to the candidate in decreasing order
#  Then find how much contribution the top-N% made
#  Then compute the fraction to the overall campaign collection
# For example, if Obama collected 1 billion dollars and the top-1% gave 100 million then their contribution is 10%

def t7a_contributions_by_top_N_pct_obama(N):
    
    fec_obama = fec_all[fec_all.cand_nm.isin(['Obama, Barack'])]
    table =  pd.pivot_table(fec_obama,values='contb_receipt_amt', index=['contbr_nm','contbr_city','contbr_zip','contbr_st','contbr_occupation','contbr_employer'],columns='cand_nm',aggfunc='sum')

    sortedtable= table.sort(columns= 'Obama, Barack',ascending=False)
    topper=table.head(1)
    totalsum=sortedtable['Obama, Barack'].sum(1)
    totalcontributer=sortedtable.count(0)
    Nper=N*totalcontributer/100
    getNtopper=sortedtable.head(int(Nper))
    output=getNtopper['Obama, Barack'].sum()/totalsum
    return output



def t7a_contributions_by_top_N_pct_romney(N):
    
    
    fec_romney = fec_all[fec_all.cand_nm.isin(['Romney, Mitt'])]
    table =  pd.pivot_table(fec_romney,values='contb_receipt_amt', index=['contbr_nm','contbr_city','contbr_zip','contbr_st','contbr_occupation','contbr_employer'],columns='cand_nm',aggfunc='sum')

    sortedtable= table.sort(columns= 'Romney, Mitt',ascending=False)
    topper=table.head(1)
    totalsum=sortedtable['Romney, Mitt'].sum(1)
    totalcontributer=sortedtable.count(0)
    Nper=N*totalcontributer/100
    getNtopper=sortedtable.head(int(Nper))
    output=getNtopper['Romney, Mitt'].sum()/totalsum
    return output


#Task 7b: Now let us see who these people in 1% are
# Compute the top-1% based on total contribution (ie. money they gave to Obama + Romney)
# Now let us see some information about them.
# For each of the folks in 1%, print the following:
#    name, state, occupation, employer, total amount they gave to Obama, total amount they gave to Romney
allTable = pd.pivot_table(fec,values='contb_receipt_amt', index=['contbr_nm','contbr_st','contbr_occupation','contbr_employer'],columns='name_cand',aggfunc='sum',fill_value=0)
allTable['total'] = allTable['Obama']+allTable['Romney']
sortedtableAll=allTable.sort(columns='total',ascending=False)
n=len(sortedtableAll)*0.01
finalTable = sortedtableAll.head(int(n))
t7b_1pcters = pd.DataFrame(finalTable)
print t7b_1pcters
```
